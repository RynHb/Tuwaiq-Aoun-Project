```python
with open('Labor_rules_v0.1.txt', encoding='UTF-8') as f:
    lines = f.readlines()
```

```python
arabic_text = ''.join(lines)
```

```python
arabic_text = arabic_text.strip()
```

```python
arabic_text
```

```python
words = word_tokenize(arabic_text)
```

```python
#tokens
```

```python
pip install Arabic-Stopwords
```

```python
import arabicstopwords.arabicstopwords as stp

filtered_tokens = [token for token in tokens if not stp.is_stop(token)]
filtered_tokens
```

```python
words
```

```python
from nltk.stem import ISRIStemmer

stemmer = ISRIStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
stemmed_tokens
```

```python
preprocessed_text = (' '.join(stemmed_tokens))
```

```python
chunks = preprocessed_text.split('باب')
formatted_text = '\nباب'.join(chunks)

with open('formatted_text.txt', 'w', encoding='utf-8') as file:
    file.write(formatted_text)
```
